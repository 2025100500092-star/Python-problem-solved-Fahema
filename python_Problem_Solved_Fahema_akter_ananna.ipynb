(c) Grade Calculator

```python
mark = int(input("Enter your mark (0-100): "))

if mark < 0 or mark > 100:
    print("Invalid mark")
elif mark >= 90:
    print("A+")
elif mark >= 80:
    print("A")
elif mark >= 70:
    print("B")
elif mark >= 60:
    print("C")
elif mark >= 50:
    print("D")
else:
    print("F - Failed")
```

(d) Login System

```python
correct_username = "admin"
correct_password = "1234"

attempts = 0

while attempts < 3:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == correct_username and password == correct_password:
        print("Login successful! Welcome,", username)
        break
    elif username == correct_username:
        print("Incorrect password. Try again.")
    else:
        print("User not found.")

    attempts += 1

if attempts == 3:
    print("Account locked.")
```

Question 5: ATM Simulation

```python
# Login first
username = "admin"
password = "1234"
is_active = True

u = input("Username: ")
p = input("Password: ")

if not is_active:
    print("Account is disabled.")
elif u == username and p == password:
    print("Login successful!")

    # ATM Section
    pinNumber = 0000
    balance = 1000
    attempts = 3

    while attempts > 0:
        pin = int(input("Enter PIN: "))

        if pin == pinNumber:
            request = int(input("Enter amount to withdraw: "))

            if request > balance:
                print("Insufficient balance.")
            elif request <= 100:
                print("Minimum withdrawal is above 500.")
            else:
                balance -= request
                print("Withdrawal successful!")
                print("Remaining balance:", balance)
            break
        else:
            attempts -= 1
            print("Wrong pin. Attempts remaining:", attempts)

    if attempts == 0:
        print("Card blocked!")
else:
    print("Login failed.")
```

Question 6: Basic Loops

```python
# 1 to 20
for i in range(1, 21):
    print(i)

# Table of 7
for i in range(1, 11):
    print("7 x", i, "=", 7 * i)

# Sum 1 to 100
total = 0
for i in range(1, 101):
    total += i
print("Sum:", total)

# Even numbers 1 to 50
for i in range(2, 51, 2):
    print(i)

# Fruits list
fruits = ['apple', 'banana', 'cherry', 'mango', 'grape']

for i in range(len(fruits)):
    print(i + 1, fruits[i])
```

(b) While Loop

```python
# 10 to 1
num = 10
while num >= 1:
    print(num)
    num -= 1


# Sum of positive numbers
total = 0
while True:
    n = int(input("Enter a positive number (0 or negative to stop): "))
    if n <= 0:
        break
    total += n

print("Total:", total)


# Countdown
count = 10
while count >= 0:
    if count == 0:
        print("Blast off!")
    else:
        print(count)
    count -= 1


# First 10 multiples of 6
count = 1
while count <= 10:
    print(6 * count)
    count += 1
```
